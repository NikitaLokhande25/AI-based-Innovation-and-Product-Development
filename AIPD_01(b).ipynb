```python
x = df.drop('label',axis=1)
y = df['label']
```

```python
x.head()
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train)
```

```python
#1) Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model = model.fit(x_train,y_train)
model
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score,f1_score

acc = accuracy_score(y_test,y_pred)
print("Accuracy : ",acc)

f1 = f1_score(y_test,y_pred, average='weighted')
print("F1 Score : ",f1)

cr = classification_report(y_test,y_pred, zero_division=0)
print("Classification Report : ",cr)

cm = confusion_matrix(y_test,y_pred)
print("Confusion Matrix : ",cm)
```

```python
sns.heatmap(cm, annot=True, fmt='g', xticklabels=["Class 0", "Class 1"], yticklabels=["Class 0", "Class 1"])
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.title("Confusion Matrix Heatmap LogisticRegression")
plt.show()
```

```python
#2) NaiveBayes Classification
## 1-- MultinomialNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler()

# Scale x_train and x_test
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

model2 = MultinomialNB()
model2 = model2.fit(x_train_scaled, y_train)
model2
```

```python
y_pred2 = model2.predict(x_test_scaled)
y_pred2
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score,f1_score

acc1 = accuracy_score(y_test,y_pred2)
print("Accuracy of Multinomial NB Model : ",acc1)

f1_n = f1_score(y_test,y_pred2, average='weighted')
print("F1 Score Multinomial NB Model : ",f1_n)

cr1 = classification_report(y_test,y_pred2, zero_division=0)
print("Classification Report Multinomial NB Model : ",cr1)

cm1 = confusion_matrix(y_test,y_pred2)
print("Confusion Matrix Multinomial NB Model : ",cm1)
```
